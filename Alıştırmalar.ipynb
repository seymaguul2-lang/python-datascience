```python
print('''**************
      Kullanıcı Girişi
      **************''')


sys_kullanici_adi = 'seyma'
sys_parola = '123'
giris_hakki = 3

while True:
    kullanici_adi = input('Kullanıcı Adınızı Giriniz:')
    Parola = input('Parolanızı Giriniz:')
    if (sys_kullanici_adi == kullanici_adi and sys_parola != Parola):
        print('Parola Hatalı')
        giris_hakki -=1
    elif (sys_kullanici_adi!= kullanici_adi and sys_parola == Parola):
        print('Kullanıcı Adı Hatalı')
        giris_hakki -=1
    elif (sys_kullanici_adi != kullanici_adi and sys_parola != Parola):
        print('Kullancı Adı ve Parola Hatalı')
        giris_hakki -=1
    else:
        print('Giriş Başarılı....')
        break
    if giris_hakki ==0:
        print('Giriş Hakkınız Bitti. Çıkış Yapılıyor....')
        break


```

```python
print('''************************** 
ATMDEN PARA ÇEKME
      1.İşlem- Bakiye Sorgulama

      2.İşlem- Para Yatırma

      3.İşlem- Para çekme

Programdan Çıkış Yapmak İçin 'q' Basınız....
      
       **************************''')

bakiye =1000

while True:
    işlem = input('İşlemi seçiniz:')
    if (işlem=='q'):
        print('Yine Bekleriz...')
        break
    elif (işlem == '1'):
        print('Bakiyeniz {} tldir.'.format(bakiye))
    
    elif (işlem=='2'):
        miktar =int(input('Paranızı Yatırabilirsiniz:'))

        bakiye += miktar

    elif (işlem == '3'):
        miktar =int(input('İstediğiniz Miktarı Girirniz.'))

        if (bakiye-miktar<0):
            print('Bakiye Yetersiz...')
            continue
        bakiye -= miktar 

    else:
        print('Geçersiz İşlem...')



```
