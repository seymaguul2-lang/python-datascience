## instance & attribute

```python
class SuperKahraman():

    ozelGuc = "Gorunmezlik"
    
    def __init__(self, isimInput, yasInput, meslekInput):
        print("init çağrıldı")
        self.isim = isimInput
        self.yas = yasInput
        self.meslek = meslekInput

    def ornekMethod(self):
        print(f"ben suparkahramanım ve mesleğim: {self.meslek}")
```

```python
superman = SuperKahraman("Superman", 30, "Gazeteci")
```

```python
superman.isim
```

```python
superman.meslek
```

```python
superman.yas
```

```python
superman.ozelGuc
```

```python
superman.ozelGuc = "Uçabiliyor"
```

```python
superman.ozelGuc
```

```python
superman.ornekMethod();
```

```python
class Kopek():

    yilCarpani = 7
    
    def __init__(self,yas=5):
        self.yas = yas

    def insanYasiniHesapla(self):
        return self.yas * self.yilCarpani
```

```python
benimKopek = Kopek(3)
```

```python
benimKopek.yas
```

```python
benimKopek.insanYasiniHesapla()
```

## inheritance

```python
class Hayvan():
    def __init__(self):
        print("hayvan sınıfı init çağırıldı.")
    def method1(self):
        print("hayvan sınıfı method1 .")
    def method2(self):
        print("hayvan sınıfı method2 çağırıldı.")
```

```python
benimHayvanim = Hayvan()
```

```python
benimHayvanim.method1()
```

```python
benimHayvanim.method2()
```

```python
class Kedi(Hayvan):
    def __init__(self):
        Hayvan.__init__(self)
        print("kedi sınıfı init çağırıldı.")

    def miyavla(self):
        print("miyav")
        
    #override
    def method1(self):
        print("Kedi sınıfındaki method1 çağırıldı.")
```

```python
benimKedi = Kedi()
```

```python
benimKedi.method1()
```

```python
benimKedi.miyavla()
```

## polymorphism

```python
class Elma():
    def __init__(self,isim):
        self.isim = isim

    def bilgiVer(self):
        return self.isim + " 100 kaloridir"
```

```python
class Muz():
    def __init__(self,isim):
        self.isim = isim

    def bilgiVer(self):
        return self.isim + " 150 kaloridir"
```

```python
elma = Elma("elma")
```

```python
elma.bilgiVer()
```

```python
muz = Muz("muz")
```

```python
muz.bilgiVer()
```

```python
meyveLİstesi = [elma,muz]
```

```python
for meyve in meyveLİstesi:
    print(meyve.bilgiVer())
```

```python
def bilgiAl(meyve):
    print(meyve.bilgiVer())
```

```python
bilgiAl(muz)
```

```python
bilgiAl(elma)
```
